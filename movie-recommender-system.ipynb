```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):         #ast.literal_eval() is used to convert string to list
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):         #ast.literal_eval() is used to convert string to list
        if cnt != 3:                        # we are taking first 3 chracters
            L.append(i['name'])
            cnt += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):         #ast.literal_eval() is used to convert string to list
        if i['job'] == 'Director':          # we are taking the director of the movie
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies['overview'][0]       # it is in string

#We will convert it to list  and then we will remove spaces from cast, crew, genres, keyowrds and append it to the list and then we will join it with space
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())       # we are converting string to list
```

```python
movies.head(1)
```

```python
# We are removing the spaces so that we can use it as a single word
# cause it might happen if any user wants to see the movies of Sam Worthinton but there it shows the movies of Sam Mendes
# to counter this problem it will take Sam WOrthinton as a single word  and will show the recommendations accoridngly

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

We have some words like  ability, abilities
then we have actor, actors
dance, dances


For these types of words we need to give only one word
ability, actor, dance

so we will applyting stemming

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

### We will vectorize the movies and we will calculate the shortest distance betweeen the movies. Technique we will use bag of words - convert text to vector

The similarity between the movies depends on how much tags of the movies ar related to one another


```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')   # this tells us that we are taking only 5000 features and we are removing the stop words
```

### We will convert the tags into vectors and remove the stop words

We will take the most common 5000 words 

We are simply taking out 5000 most frequent words

```python
vectors = cv.fit_transform(new_df['tags']).toarray()  # it will convert the tags into vectors 
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

We will take the cosine distance, not the euclidean distance

For higher dimension, euclidean distance is not a good measure

euclidean distance gives the distance tip to tip but in cosine distance, it gives the angle difference

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        # print(i[0])
        print(new_df.iloc[i[0]].title)
```
